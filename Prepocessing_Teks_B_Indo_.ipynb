Mengimpor Teks dari Web


```python
import requests
from bs4 import BeautifulSoup
```

```python
results = requests.get('https://warstek.com/bepicolombo/')
results.text
```

```python
# Encoding
results.encoding = results.apparent_encoding
results.encoding
```

```python
WP = BeautifulSoup(results.text, 'html.parser')
```

```python
# Mendefinisikan seluruh teks, menghilangkan tag

fullteks = []

for paragraf in WP.find_all('p'):
    fullteks.append(paragraf.getText())
    
fullteks
```

```python
# Menghilangkan white space
fullteks = [paragraf.strip() for paragraf in fullteks if paragraf != '']
fullteks
```

```python
# Menggabungkan Tiap Line

fullteks = ' '.join(fullteks)
fullteks
```

Preprocessing Teks

```python
import re
import nltk
from nltk import word_tokenize
nltk.download('punkt')
nltk.download('stopwprd')
nltk.download('wordnet')
```

```python
teks = fullteks
```

```python
# Mengubah hurufnya jadi lower semua
teks = teks.lower()
teks
```

```python
# Menghilangkan karakter2
teks = re.sub('r[-()\"#/@;:<>{}+=|!?,]','', teks)
teks   
```

```python
# Tokenize
token = word_tokenize(teks)
print(token)
```
